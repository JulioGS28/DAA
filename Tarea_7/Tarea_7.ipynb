```python
archivo = open('el_quijote_ii.txt',mode='r')
texto = archivo.read()


def string_match_brute(A,P):
  con = [] # Lista que nos ayudara a almacenar el total de palabras en el texto
  for c in range (len(P)):
    contador = 0
    P1 = P[c]
    for i in range( len(A) - len(P1) + 1):
      for j in range(len(P1)):
        if P1[j] == A[i+j]:
          pass
        else:
          break
      if j+1 == len(P1) and A[i+j] == P1[j]:
        contador = contador + 1
        
        print(f"{P1}: Se encontro en la posicion {i}")
        
    con.append(contador)
    
  for e in range(len(P)): 
    print(f"---{P[e]}: se encontro {con[e]} veces en el texto---")
  
entrada = texto
patron=["hidalgo","molino","merced"]
string_match_brute(entrada,patron)
```
